```python
import numpy as np
import pandas as pd
import os
```

```python
article_read = pd.read_csv(r'C:\Users\g702726\Desktop\aa\bank-additional\bank-additional-full.csv', delimiter=';')
article_read
```

```python
article_dictionary = article_read.T.to_dict().values()
#article_dictionary
```

```python
from sklearn.feature_extraction import DictVectorizer
vec = DictVectorizer()
A = vec.fit_transform(article_dictionary).toarray()
B = vec.get_feature_names()
#X,Y
```

```python
article=pd.DataFrame(A, columns = B)
article
b=['y=no','y=yes']
a= [i for i in B if i not in b]
x = A[:,:-2]
#data = pd.DataFrame(A, columns = a)
#data
X = np.hstack((x[:,:14],x[:,15:]))
y = A[:,-1]
article
```

```python
import matplotlib.pyplot as plt
#Random Forest Classification Algorithm to get feature importance
#this part is directly taken from "https://scikit-learn.org/stable/auto_examples/ensemble/plot_forest_importances.html"

from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier


forest = RandomForestClassifier(n_estimators=250,random_state=0)

forest.fit(X, y)
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(X.shape[1]):
    print("%d. feature %s (%f)" % (f + 1, B[indices[f]], importances[indices[f]]))

# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(X.shape[1]), importances[indices], color="b", yerr=std[indices], align="center")
plt.xticks(range(X.shape[1]), indices)
plt.xlim([-1, X.shape[1]])
plt.show()
```

```python
#Optimization on the number of calls
#Success Ratio
total_success= sum(article['y=yes'])
total_campaign=sum(article['campaign'])
success_ratio= total_success/total_campaign
print(success_ratio)
```

```python
print("Call Number  \t Efficiency")
for i in range(1,24):
    z=article.loc[article['campaign']==i]
    u=article.loc[article['campaign']>=i]
    s=sum(z['y=yes'])
    t=u.shape[0]
    efficiency = s/t
    print(str(i)+"\t\t"+str(efficiency))
    
#As the conversion ratio is less than 2% after 6th call, so there is no point to call any customer after 6th call.
```

```python
#Optimizing in calls
total_number_of_calls= sum(article['campaign'])#total number of calls
print("Total Number of calls:  " + str(total_number_of_calls))
i= sum(article[article['campaign']>6]['campaign'])#total number of calls for campaign greater than 6
j= article[article['campaign']>6]['campaign'].shape[0] #calculation of the total number of minimum calls of 6 to a customer
extra_calls=i-(6*j)
print("Extra calls:  " + str(extra_calls))
Reduction = (extra_calls/total_number_of_calls)*100 #Reduction in marketing cost
print("Reduction in marketing in percentage:  "+str(Reduction))
```
